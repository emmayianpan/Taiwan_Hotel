```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
#Read the ods File
path = "../static/data/2019/combine19.csv"
df = pd.read_csv(path)
df
```

```python
area_room = df.groupby(['Area','Date']).sum()['Room']
area_supply = df.groupby(['Area','Date']).sum()['Supply']
area_demand = df.groupby(['Area','Date']).sum()['Demand']
area_rmrev = df.groupby(['Area','Date']).sum()['Room_Revenue']
area_fbrev = df.groupby(['Area','Date']).sum()['F&B_Revenue']
area_ttlrev = df.groupby(['Area','Date']).sum()['Total_Revenue']
```

```python
#Calculate Monthly Occupancy/ADR/RevPAR
area_occ = round((area_demand/area_supply),2)
area_adr = round((area_rmrev/area_demand), 2)
area_revpar = round((area_rmrev/area_supply), 2)
```

```python
area_summary = pd.DataFrame({
    'Room': area_room,
    'Supply': area_supply,
    'Demand': area_demand,
    'Occupancy': area_occ,
    'ADR': area_adr,
    'RevPAR': area_revpar,
    'Room_Revenue': area_rmrev, 
    'F&B_Revenue': area_fbrev,
    'Total_Revenue': area_ttlrev  
})

area_summary
```
